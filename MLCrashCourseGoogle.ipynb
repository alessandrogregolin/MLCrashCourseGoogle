# **Basic terminology**






**Key basic terminology**
*   **Labels**: what we are trying to predict (y)
*   **Features**: input variables {x1, x2, x3 ...}
*   **Examples**: particular instances of data. **Labeled examples** contain both features and the label (used to train the model). An **unlabeled example** only contains features (used to predict the label)
* **Model** defines the relationship between features and label. Training means creating or learning the model. Inference means applying the trained model to unlabeled examples.
* **Regression model**  predicts continuous values such as prices or probability
* **Classification** predicts discrete values such as T or F


# **Linear Regression and Loss**



Linear regression is a method for finding the straight line or hyperlane that best fits a set of pointsSchermata 2021-04-29 alle 23.30.29.png

In this case there is a linear relationship that can be represented in machine learning by the equation


```
y' = b + w1*x1
```
Where

 * y' is the predicted label (a desired output).
 * b is the bias (the y-intercept), sometimes referred to as w0
 * w1 is the weight of feature 1. Weight is the same concept as the "slope" 
 in the traditional equation of a line.

 * x1 is a feature (a known input).

 To infer (predict) the label y' we just have to substitute x1 in the model. a more complicated model might have various features x1...xN with corresponding weights w1...wN to infer a label y'


**Training** a model means learning good values for all the weights and the bias from the labeled example. In supervised machine learning the algorithm builds a model examining many examples and attempts to find a model that minimizes loss.

**Loss** is a number indicating how bad the model's prediction was on a single example. The larger the loss the worse the predicition. The goal is to find weights and biases that have low loss, on average, across all examples.Schermata 2021-04-29 alle 23.41.55.png

**Squared loss (also called L2 Loss, in Stats we call it error sum of squares)** is a popular loss function and is the difference between the label and the prediction squared

```
= (observation - prediction(x))^2
= (y - y')^2
```

**Mean square error** is the average square loss per example over the whole dataset. It is calculated by summing up all the square losses for individual example and dividing for the number of examples.

Using a **linear regression model** as an example --> We can start making a prediction using random values for w0(b) and for w1. We could for example take 0 and 0. We then use the **loss function** to compute the loss for the prediction made using these two parameters --> suppose we choose to use the **squared loss function**. The result of this function is then passed to the **compute parameter updates part** where new and better values for w0 and w1 are computed. The entire process then is repeated until the algorithm finds the values for which the loss is minimised. The result is that the model has **converged**.
